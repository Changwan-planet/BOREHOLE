```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.ticker import MaxNLocator
import numpy as np


#Define Matplotlib figure and axis
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()


#add rectangle to plot
#add the borehole stratigraphy of section
#Rectangle((x and y), wide, height, facecolor="", fill=True, patch="+,//,\\,-")

#x and y means lower left point of the rectangle.
#height means the depth of the bohore holes.


#Stratigraphy function
#input values ((x,y),depth,Patch)
#apparent depth

#     ++++++++++++++++
#++++++rectangle wide++++++
#     ++++++++++++++++
w=6

#Assum the borehole wide for convenience (not scale)
b_w=w/2.0

#     ++++++++++++++++++++++
#++++++Longitudial distance++++++
#     ++++++++++++++++++++++
#Data read from DEM
x = [10, 20, 30, 50] 

#     ++++++++++++++++++++++++++
#++++++Mormology of the section++++++
#     ++++++++++++++++++++++++++

#Data read from DEM
elev=[76.28, 76.42, 76.74, 76.72]

#     +++++++++++++++++++++++
#++++++Depth of the Borehole++++++
#     +++++++++++++++++++++++


#Data read from the Borehole analysis
#Borehole_a
a_d = [0.8, 1.7, 0.5, 1.5, 1.5]

#Borehole_b
b_d = [0.5, 0.5, 0.2, 0.8, 
       1.0, 0.2, 0.3, 2.5, 
       2.0, 1.3]

#Borehole_c
c_d = [0.4, 0.6, 1.0, 0.7,
       0.3, 0.7, 0.3, 1.8,
       0.7, 0.5, 0.35, 0.65]

#Borehole_d
d_d = [0.4, 0.6, 1.0, 1.2,
       0.8, 0.5, 2.5]

#     +++++++++++++++++++++++++++++++++++++
#++++++Lower left y point of the rectangle++++++
#     +++++++++++++++++++++++++++++++++++++
#a
n = len(a_d)
a_z = [0 for i in range(n)]
for i in range(n):
 a_z[i] = elev[0] - sum(a_d[:i+1])

#b
n = len(b_d)
b_z = [0 for i in range(n)]
for i in range(n):
 b_z[i] = elev[1] - sum(b_d[:i+1])

#c
n = len(c_d)
c_z = [0 for i in range(n)]
for i in range(n):
 c_z[i] = elev[2] - sum(c_d[:i+1])

#d
n = len(c_d)
d_z = [0 for i in range(n)]
for i in range(n):
 d_z[i] = elev[2] - sum(d_d[:i+1])

#     +++++++++++++++++
#++++++Draw the graphs++++++
#     +++++++++++++++++

#     +++++++++++++++++++++++++++
#++++++Morphology of the section++++++
#     +++++++++++++++++++++++++++
elev_x=x[:]
elev_z=elev[:]
ax1.plot(elev_x,elev_z,'black')


#      +++++++++++++++++++++++++++++
#+++++++Stratigraphy of the section++++++
#      +++++++++++++++++++++++++++++
#layer_1
layer_1_x = x[:]
layer_1_z = [a_z[0], b_z[1], c_z[1], d_z[1]]
#layer_1_y=[a_y_1, b_y_1, c_y_1, d_y_1]

#layer_2
layer_2_x = x[:]
layer_2_z = [a_z[1], b_z[4], c_z[3], d_z[3]]

#black mud layer
layer_md_x = x[2:]
layer_md_z = [c_z[2], d_z[2]]

#basement layer
layer_bs_x = x[:]
layer_bs_z = [a_z[2], b_z[6], c_z[6], d_z[3]]


ax1.plot(layer_1_x,layer_1_z,'saddlebrown',lw=3)
#ax1.plot(layer_2_x,layer_2_z,'olive')
ax1.plot(layer_md_x,layer_md_z,'black',lw=3)
ax1.plot(layer_bs_x,layer_bs_z,'grey',lw=3)

plt.text(layer_bs_x[0]-1.7*b_w,3.5, str(elev[0]-layer_bs_z[0]), size=20, fontweight="bold")
plt.text(layer_bs_x[1]-1.7*b_w,3.9, str(elev[1]-layer_bs_z[1]), size=20, fontweight="bold")
plt.text(layer_bs_x[2]-1.7*b_w,4.1, str(elev[2]-layer_bs_z[2]), size=20, fontweight="bold")
plt.text(layer_bs_x[3]-1.7*b_w,3.3, str(round(elev[3]-layer_bs_z[3],1)), size=20, fontweight="bold")

#ax2.plot(layer_3_x,layer_3_y,'black')
         
   
    
#     ++++++++++
#++++++Borehole++++++
#     ++++++++++
"""
def bh_str(*x_k,  *xk_d_1,  *xk_elev, *colors):
   w = 4    
    b_w = w/2.0
    z_k = xk_elev - xk_d_1

    ax1.add_patch(Rectangle((x_k-b_w, z_k), w, xk_d_1, fill = True, facecolor = colors)                            
bh_str(10,  0.5,  76.7, "brown")
"""
                            
#Borehole_a
a_1 = 0
a_2 = 1
a_3 = 2
a_4 = 3
a_5 = 4

ax1.add_patch(Rectangle((x[a_1]-b_w, a_z[a_1]), w, a_d[a_1], fill = True,
                        facecolor = 'saddlebrown', hatch = '.....'))
ax1.add_patch(Rectangle((x[a_1]-b_w, a_z[a_2]), w, a_d[a_2], fill = True,
                        facecolor = 'yellow', hatch = '.....'))
ax1.add_patch(Rectangle((x[a_1]-b_w, a_z[a_3]), w, a_d[a_3], fill = True,
                        facecolor = 'olive', hatch = 'O'))
ax1.add_patch(Rectangle((x[a_1]-b_w, a_z[a_4]), w, a_d[a_4], fill = True,
                        edgecolor="black",lw = 3, facecolor = 'none', hatch='/O'))
ax1.add_patch(Rectangle((x[a_1]-b_w, a_z[a_5]), w, a_d[a_5], fill = True,
                        facecolor = 'lightgrey', hatch = '//'))


#Borehole_b
ax1.add_patch(Rectangle((x[1]-b_w, b_z[0]), w, b_d[0], fill = True, 
                        edgecolor = 'black', lw=3,facecolor = 'none'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[1]), w, b_d[1], fill = True, 
                        facecolor = 'saddlebrown', hatch = '.....'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[2]), w, b_d[2], fill = True, 
                        edgecolor = 'black', lw=3,facecolor = 'none'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[3]), w, b_d[3], fill = True, 
                        facecolor = 'darkgoldenrod', hatch = 'O'))

ax1.add_patch(Rectangle((x[1]-b_w, b_z[4]), w, b_d[4], fill = True, 
                        facecolor = 'olive', hatch = 'O'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[5]), w, b_d[5], fill = True, 
                        edgecolor = 'black', lw=3,facecolor = 'none'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[6]), w, b_d[6], fill = True, 
                        facecolor = 'goldenrod', hatch = 'O'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[7]), w, b_d[7], fill = True, 
                        facecolor = 'silver', hatch = 'xx'))

ax1.add_patch(Rectangle((x[1]-b_w, b_z[8]), w, b_d[8], fill = True, 
                        edgecolor = 'black', lw=3,facecolor = 'none'))
ax1.add_patch(Rectangle((x[1]-b_w, b_z[9]), w, b_d[9], fill = True, 
                        facecolor = 'dimgrey', hatch = '//'))


#Borehole_c
ax1.add_patch(Rectangle((x[2]-b_w, c_z[0]), w, c_d[0], fill = True, 
                        edgecolor = 'black', lw=3,facecolor = 'none'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[1]), w, c_d[1], fill = True,
                        facecolor = 'saddlebrown', hatch = '.....'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[2]), w, c_d[2], fill = True,
                        facecolor = 'black', hatch = ''))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[3]), w, c_d[3], fill = True,
                        facecolor = 'olive', hatch = 'O'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[4]), w, c_d[4], fill = True, 
                        edgecolor = 'black', lw=3, facecolor = 'none', hatch = 'O'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[5]), w, c_d[5], fill = True, 
                        facecolor = 'darkgoldenrod', hatch = 'O'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[6]), w, c_d[6], fill = True, 
                        edgecolor = 'black', lw=3, facecolor = 'none', hatch = 'xxO'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[7]), w, c_d[7], fill = True, 
                        facecolor = 'dimgrey', edgecolor='maroon',hatch = 'xx'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[8]), w, c_d[8], fill = True, 
                        edgecolor = 'black', lw=3, facecolor = 'none', hatch = ''))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[9]), w, c_d[9], fill = True, 
                        facecolor = 'dimgrey', edgecolor='maroon',hatch = 'xx'))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[10]), w, c_d[10], fill = True, 
                        edgecolor = 'black', lw=3, facecolor = 'none', hatch = ''))
ax1.add_patch(Rectangle((x[2]-b_w, c_z[11]), w, c_d[11], fill = True, 
                        facecolor = 'dimgrey', hatch = 'xx'))


#Borehole_d
ax1.add_patch(Rectangle((x[3]-b_w, d_z[0]), w, d_d[0], fill = True,
                        edgecolor= 'black', lw=3,facecolor = 'none'))
ax1.add_patch(Rectangle((x[3]-b_w, d_z[1]), w, d_d[1], fill = True,
                        facecolor = 'saddlebrown', hatch = '.....'))
ax1.add_patch(Rectangle((x[3]-b_w, d_z[2]), w, d_d[2], fill = True,
                        facecolor = 'black', hatch = ''))
ax1.add_patch(Rectangle((x[3]-b_w, d_z[3]), w, d_d[3], fill = True,
                        facecolor = 'olive', hatch = 'O'))
ax1.add_patch(Rectangle((x[3]-b_w, d_z[4]), w, d_d[4], fill = True,
                        facecolor = 'lightgrey', hatch = 'xx'))
ax1.add_patch(Rectangle((x[3]-b_w, d_z[5]), w, d_d[5], fill = True,
                        facecolor = 'dimgrey',edgecolor='maroon', hatch = 'xx'))
ax1.add_patch(Rectangle((x[3]-b_w, d_z[6]), w, d_d[6], fill = True,
                        facecolor = 'dimgrey', hatch = 'xx'))

#Limits of x and y
#ax1.set_ylim(0,max_d)
#ax1.set_ylim(67.74,76.74)
ax1.set_ylim(max(elev)-9.3,max(elev))
ax2.set_ylim(0,9.3)


#test plot
#ax1.plot([0,50],[76.24,76.24])
#ax1.plot([0,50],[76.7,76.7])


#invert the y axis
ax2.invert_yaxis()

#Labels of x and y
ax1.set_xlabel('Easting [m]', size='30')
ax1.set_ylabel('Elevation [m]',size='30')
ax2.set_ylabel('Apparent depth [m]',size='30', labelpad=10)


#Ticks and ticklabels
#plt.xticks(fontsize="20")
ax1.set_yticks([max(elev),max(elev)-9.3])
ax1.set_yticklabels(["76.74", "67.44"],size='20')

ax1.tick_params(labelsize = 20)
ax2.tick_params(labelsize = 20)

#Axis and grid
#ax1.yaxis.set_major_locator(MaxNLocator(1))
#ax1.grid(True)
#ax1.grid(True, which="both", color="grey", linewidth=1, linestyle="--")

ax2.grid(True)
ax2.yaxis.set_major_locator(MaxNLocator(10))
ax2.grid(True, which="both", color="grey", linewidth=1, linestyle="--")

#Figure size
plt.rcParams["figure.figsize"] = (20, 10)

plt.title("T_221124_USudm_Western_Section", size ="20")



#plt.text(x[0]-b_w,-0.5,"South",size=20)
#plt.text(x[3]-b_w,-0.5,"North",size=20)

#plt.text(x[2]-b_w,-0.5,"U_1",size=20)
#plt.text(x[0]-b_w,-0.5,"BH_211007_2",size=20)
#plt.text(x[1]-b_w,-0.5,"BH_211007_3",size=20)
#plt.text(x[3]-b_w,-0.5,"BH_211007_4",size=20)

fig.savefig("T_221124_USudm_Western_Section.png")
#plt.show()
```
