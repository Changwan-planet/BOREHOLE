```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.ticker import MaxNLocator
import numpy as np


#add rectangle to plot
#add the borehole stratigraphy of section
#Rectangle((x and y), wide, height, facecolor="", fill=True, patch="+,//,\\,-")

#x and y means lower left point of the rectangle.
#height means the depth of the bohore holes.


#Stratigraphy function
#input values ((x,y),depth,Patch)
#apparent depth

#     ++++++++++++++++
#++++++rectangle wide++++++
#     ++++++++++++++++
w=3

#Assum the borehole wide for convenience (not scale)
b_w=w/2.0

print("b_w=",b_w)

#     ++++++++++++++++++++++
#++++++Longitudial distance++++++
#     ++++++++++++++++++++++
#Data read from DEM
x = [5, 10, 15, 20] 

#     ++++++++++++++++++++++++++
#++++++Mormology of the section++++++
#     ++++++++++++++++++++++++++

#Data read from DEM
elev=[3, 3, 3, 3]

#     +++++++++++++++++++++++
#++++++Depth of the Borehole++++++
#     +++++++++++++++++++++++


#Data read from the Borehole analysis
#Borehole_a
a_d = [0.4, 0.55, 0.25, 0.2, 0.7]

#Borehole_b
b_d = [0.4, 0.1, 0.3, 0.15, 0.20,
       0.95]

#Borehole_c
c_d = [0.25, 0.25, 0.1, 0.6, 1.1]

#Borehole_d
d_d = [0.6, 0.6, 1.00, 0.3]


#     +++++++++++++++++++++++++++++++++++++
#++++++Lower left y point of the rectangle++++++
#     +++++++++++++++++++++++++++++++++++++
#a
n = len(a_d)
a_z = [0 for i in range(n)]
for i in range(n):
     a_z[i] = elev[0] - sum(a_d[:i+1])

#b
n = len(b_d)
b_z = [0 for i in range(n)]
for i in range(n):
     b_z[i] = elev[1] - sum(b_d[:i+1])

#c
n = len(c_d)
c_z = [0 for i in range(n)]
for i in range(n):
     c_z[i] = elev[2] - sum(c_d[:i+1])

#d
n = len(d_d)
d_z = [0 for i in range(n)]
for i in range(n):
    d_z[i] = elev[3] - sum(d_d[:i+1])


print(a_z)
print(b_z)
print(c_z)
print(d_z)
    
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, axs = plt.subplots(5, 1, constrained_layout=True, figsize=(10, 15))


hatches =     ['...', 
               
               '',
               '...**',
                                      
               '...+',
               
                              
               'xx']

facecolors =  ['saddlebrown',
               
               'olive',
               'darkgoldenrod',
                             
               'goldenrod', 
               
               
               'black']

description = ['topsoil',
               
               'clay', 
               
               'sand + subround or subangular pebble',
               
               'sand + angular couble (4th layer) ', 
               
               'basement']

edgecolors =  ['black', 
               
               'black',
               
               'black',
               
               'black',
               
               'white']


def hatches_plot(ax, d, fc, ec, h):
    ax.add_patch(Rectangle((0,0), 2, 1, fill=True, facecolor=fc, edgecolor= ec, lw = 1, hatch=h))
    ax.text(3.5, 0.5, f" {d} ", size=15, ha="center")
    ax.axis('equal')
    ax.axis('off')

for ax, d, fc, ec, h in zip(axs.flat, description, facecolors, edgecolors, hatches):
    hatches_plot(ax,d, fc, ec, h)
    
fig.savefig("T_221123_USudm_Eastern_Section_texture.jpg")
```
