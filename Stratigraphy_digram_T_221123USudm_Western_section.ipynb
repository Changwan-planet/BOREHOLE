```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.ticker import MaxNLocator
import numpy as np


#add rectangle to plot
#add the borehole stratigraphy of section
#Rectangle((x and y), wide, height, facecolor="", fill=True, patch="+,//,\\,-")

#x and y means lower left point of the rectangle.
#height means the depth of the bohore holes.


#Stratigraphy function
#input values ((x,y),depth,Patch)
#apparent depth

#     ++++++++++++++++
#++++++rectangle wide++++++
#     ++++++++++++++++
w=3

#Assum the borehole wide for convenience (not scale)
b_w=w/2.0

print("b_w=",b_w)

#     ++++++++++++++++++++++
#++++++Longitudial distance++++++
#     ++++++++++++++++++++++
#Data read from DEM
x = [5, 10, 15, 20, 25] 

#     ++++++++++++++++++++++++++
#++++++Mormology of the section++++++
#     ++++++++++++++++++++++++++

#Data read from DEM
elev=[3, 3, 3, 3, 3]

#     +++++++++++++++++++++++
#++++++Depth of the Borehole++++++
#     +++++++++++++++++++++++


#Data read from the Borehole analysis
#Borehole_a
a_d = [0.3, 0.05, 0.35, 0.3, 0.4,
       1.0, 0.2]

#Borehole_b
b_d = [0.3, 0.6, 0.4, 0.85, 0.35]

#Borehole_c
c_d = [0.3, 0.5, 1.2, 0.4]

#Borehole_d
d_d = [0.4, 0.4, 0.25, 0.05, 0.25,
       0.8, 0.2]

#Borehole_
e_d = [0.25, 0.55, 0.2, 0.4, 0.8]


#     +++++++++++++++++++++++++++++++++++++
#++++++Lower left y point of the rectangle++++++
#     +++++++++++++++++++++++++++++++++++++
#a
n = len(a_d)
a_z = [0 for i in range(n)]
for i in range(n):
     a_z[i] = elev[0] - sum(a_d[:i+1])

#b
n = len(b_d)
b_z = [0 for i in range(n)]
for i in range(n):
     b_z[i] = elev[1] - sum(b_d[:i+1])

#c
n = len(c_d)
c_z = [0 for i in range(n)]
for i in range(n):
     c_z[i] = elev[2] - sum(c_d[:i+1])

#d
n = len(d_d)
d_z = [0 for i in range(n)]
for i in range(n):
    d_z[i] = elev[3] - sum(d_d[:i+1])

#e
n = len(e_d)
e_z = [0 for i in range(n)]
for i in range(n):
    e_z[i] = elev[4] - sum(e_d[:i+1])


print(a_z)
print(b_z)
print(c_z)
print(d_z)
print(e_z)    
    
```
