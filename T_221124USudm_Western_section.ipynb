```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, axs = plt.subplots(7, 1, constrained_layout=True, figsize=(5, 10))


hatches =     ['...', 
               '...**',
               '...',
               '...**',
               '',
               '...+',
                                      
               
                              
               'xx']

facecolors =  ['saddlebrown',
               'saddlebrown',
               'goldenrod',     
               'goldenrod', 
               'goldenrod',
               'goldenrod',
               
               'black']

description = ['topsoil',
               'topsoil + pebble',
               'sand', 
               'sand + granule ~ pebble',
               'clay',
               '4th layer',
               
               'basement']

edgecolors =  ['black', 
               'black',
               'black',
               'black',
               'black',
               'black',
                              
               'white']


def hatches_plot(ax, d, fc, ec, h):
    ax.add_patch(Rectangle((0,0), 2, 1, fill=True, facecolor=fc, edgecolor= ec, lw = 1, hatch=h))
    ax.text(4.5, 0.5, f" {d} ", size=15, ha="center")
    ax.axis('equal')
    ax.axis('off')

for ax, d, fc, ec, h in zip(axs.flat, description, facecolors, edgecolors, hatches):
    hatches_plot(ax,d, fc, ec, h)

fig.savefig("T_221124_USudm_Western_Section_texture.jpg")
```
