```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, axs = plt.subplots(9, 1, constrained_layout=True, figsize=(9, 12))


hatches =     ['',   '...',  '.' , 

               '*',  '...*',
               
               '**', '...**',  '...+', 
                                          
               'xx']

edgecolors =  ['black', 'black', 'black', 
               
               'black', 'black',  
               
               'black', 'black', 'black',  
               
               'black']

description = ['clay',   'silt',     'sand',  

               'clay + pebble~cobble', 'sand + pebble~cobble',

               'clay + granule~pebble',  'sand + granule~pebble', 'sand + angular cobble',
               
               'fractured basement']

def hatches_plot(ax, d, ec, h):
    ax.add_patch(Rectangle((0,0), 2, 1, fill=False, edgecolor= ec, lw = 1, hatch=h))
    ax.text(3.5, 0.5, f" {d} ", size=15, ha="center")
    ax.axis('equal')
    ax.axis('off')

for ax, d, ec, h in zip(axs.flat, description, edgecolors, hatches):
    hatches_plot(ax, d, ec, h)
    
fig.savefig("texture.jpg")
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, axs = plt.subplots(3, 1, constrained_layout=True, figsize=(4, 4))


hatches =     ['...', 
               '...+',
               ''
                ]

facecolors =  ['saddlebrown',
               'goldenrod',
               'black'
                ]

description = ['topsoil',
               '4th layer',
               'black line'
               ]

edgecolors =  ['black', 
               'black',
               'black']


def hatches_plot(ax, d, fc, ec, h):
    ax.add_patch(Rectangle((0,0), 2, 1, fill=True, facecolor=fc, edgecolor= ec, lw = 1, hatch=h))
    ax.text(3.5, 0.5, f" {d} ", size=15, ha="center")
    ax.axis('equal')
    ax.axis('off')

for ax, d, fc, ec, h in zip(axs.flat, description, facecolors, edgecolors, hatches):
    hatches_plot(ax,d, fc, ec, h)
    
fig.savefig("Special_section.jpg")
```
